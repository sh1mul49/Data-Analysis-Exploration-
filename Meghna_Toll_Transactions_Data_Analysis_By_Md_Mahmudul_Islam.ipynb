```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df =pd.read_excel('Data.xlsx')
df.head()
```

```python
df= df.drop(columns=["STATUS_REMARKS", "UPDATED_AT", "CANCELED_AT", "CANCELED_BY"])

df["CREATED_AT"] = pd.to_datetime(df["CREATED_AT"], format="%d-%b-%y %I.%M.%S.%f %p")
```

Time series plot of transactions over time

```python

df['transaction_date'] = df['CREATED_AT'].dt.date
transactions_over_time = df.groupby('transaction_date').size().reset_index(name='transaction_count')
transactions_over_time['transaction_date'] = pd.to_datetime(transactions_over_time['transaction_date'])

plt.figure(figsize=(15, 7))
sns.lineplot(x='transaction_date', y='transaction_count', data=transactions_over_time)
plt.title('Number of Transactions Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Distribution of AMOUNT

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["AMOUNT"], bins=50, kde=True)
plt.title("Distribution of Transaction Amount")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

Vehicle Class Distribution

```python
plt.figure(figsize=(12, 7))
sns.countplot(data=df, y="VEHICLE_CLASS", order=df["VEHICLE_CLASS"].value_counts().index, palette="viridis")
plt.title("Distribution of Vehicle Classes")
plt.xlabel("Count")
plt.ylabel("Vehicle Class")
plt.tight_layout()
plt.show()
```

Transactions by Hour of Day

```python
# Convert 'CREATED_AT' to datetime and extract time features
df["CREATED_AT"] = pd.to_datetime(df["CREATED_AT"])
df["HOUR_OF_DAY"] = df["CREATED_AT"].dt.hour
df["DAY_OF_WEEK"] = df["CREATED_AT"].dt.day_name()

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x="HOUR_OF_DAY", palette="magma")
plt.title("Number of Transactions by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Transactions")
plt.show()
```

Transactions by Day of Week

```python

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='DAY_OF_WEEK', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette='cividis')
plt.title('Number of Transactions by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Transactions')
plt.show()

```

Distributions of Some Attributes

```python
cat_preview_cols = ['VEHICLE_CLASS','LANE_NO','BANK_ID','PAYMENT_TYPE_ID','PLAZA_ID','IS_ACTIVE','CLASS_STATUS']
for c in cat_preview_cols:
    if c in df.columns:
        print(f"\nTop value counts for {c}:")

        print(df[c].value_counts(dropna=False).head(20))

        plt.figure(figsize=(10, 5))
        df[c].value_counts(dropna=False).plot(kind='bar')
        plt.title(f"Value Counts for {c}")
        plt.xlabel(c)
        plt.ylabel("Count")
        plt.show()
```

Lane wise vechicle count

```python
plt.figure(figsize=(15, 8))
sns.countplot(data=df, x='LANE_NO', hue='VEHICLE_CLASS', palette='tab20')
plt.title('Count of Vehicle Class by Lane Number')
plt.xlabel('LANE_NO')
plt.ylabel('Count')
plt.legend(title='VEHICLE_CLASS', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
```

Vehicle Class by Recognize By

```python
plt.figure(figsize=(18, 8))
sns.countplot(data=df, x='RECOGNIZE_BY', hue='VEHICLE_CLASS', palette='tab20')
plt.title('Count of Vehicle Class by Recognize By')
plt.xlabel('Recognize By')
plt.ylabel('Count')
plt.legend(title='VEHICLE_CLASS', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
```

```python
traffic_data = df.groupby(["LANE_NO", "HOUR_OF_DAY", "VEHICLE_CLASS"]).size().reset_index(name="Vehicle_Count")

g = sns.FacetGrid(traffic_data, col="LANE_NO", col_wrap=3, height=4, aspect=1.2, sharey=True)

g.map_dataframe(sns.lineplot, x="HOUR_OF_DAY", y="Vehicle_Count", hue="VEHICLE_CLASS", palette="tab10")

g.set_axis_labels("Hour of Day", "Vehicle Count")
g.set_titles(col_template="LANE_NO={col_name}")
g.add_legend(title="VEHICLE_CLASS", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.suptitle("Peak Time Traffic of Vehicle Classes per Lane", y=1.02)

plt.tight_layout()
```

Heatmap

```python

heatmap_data = df.groupby(["VEHICLE_CLASS", "HOUR_OF_DAY"]).size().unstack(fill_value=0)

plt.figure(figsize=(16, 9))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="Blues", linewidths=.5)
plt.title("Vehicle Class Distribution Across Hours (All Lanes Combined)")
plt.xlabel("Hour of Day")
plt.ylabel("Vehicle Class")
plt.tight_layout()
```
